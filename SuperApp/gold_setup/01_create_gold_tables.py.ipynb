# Gold Layer Setup - Tablas Dimensionales
Este notebook crea las tablas del gold layer:
1. gold_categorias (dimension table)
2. gold_productos_categorias (mapping table)
3. Seeds iniciales de categorías

```python
%sql
-- Crear gold_categorias (dimension table jerárquica)
CREATE TABLE IF NOT EXISTS workspace.superapp.gold_categorias (
  id_categoria BIGINT NOT NULL,
  nombre STRING NOT NULL,
  nivel STRING NOT NULL,
  parent_id BIGINT,
  descripcion STRING,
  fecha_creacion TIMESTAMP
) COMMENT 'Tabla dimensional de categorías y subcategorías de productos';
```
